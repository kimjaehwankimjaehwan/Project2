```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# 1. 문자형 컬럼 선택 및 인코딩
merged_cat = merged.select_dtypes(include=['object'])
merged_cat = merged_cat.fillna('Missing')
merged_cat_encoded = pd.get_dummies(merged_cat)

# 2. 수치형 컬럼 선택 및 결측치 처리
merged_num = merged.select_dtypes(exclude=['object'])
num_imputer = SimpleImputer(strategy='mean')
merged_num_imputed = pd.DataFrame(num_imputer.fit_transform(merged_num), columns=merged_num.columns)

# 3. 수치형 정규화
scaler = StandardScaler()
merged_num_scaled = pd.DataFrame(scaler.fit_transform(merged_num_imputed), columns=merged_num.columns)

# 4. 결합
df_final = pd.concat([merged_num_scaled, merged_cat_encoded], axis=1)
```

```python
y_column_name = 'Followers'
X = df_final.drop(y_column_name, axis=1)
y = df_final[y_column_name]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

model = RandomForestRegressor()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model.fit(X_train, y_train)

train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print(f"R2(Train / Test) : {r2_score(y_train, train_pred):.4f} / {r2_score(y_test, test_pred):.4f}")
print(f"RMSE(Train / Test) : {mean_squared_error(y_train, train_pred)**0.5:,.0f} / {mean_squared_error(y_test, test_pred)**0.5:,.0f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))
sns.scatterplot(x=y_test, y=test_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # reference line
plt.xlabel("Actual Values (y_test)")
plt.ylabel("Predicted Values (test_pred)")
plt.title("Actual vs Predicted")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
residuals = y_test - test_pred

plt.figure(figsize=(6,4))
sns.histplot(residuals, bins=30, kde=True)
plt.title("Residual Distribution")
plt.xlabel("Residuals (Actual - Predicted)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 모델 훈련 이후
importances = model.feature_importances_
features = X_train.columns

# 중요도를 데이터프레임으로 정리
feat_importances = pd.Series(importances, index=features)

# 상위 20개 피처 시각화 (내림차순 정렬)
plt.figure(figsize=(10, 6))
sns.barplot(x=feat_importances.nlargest(20), y=feat_importances.nlargest(20).index)
plt.title("Top 20 Feature Importances (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
# 모델의 구조를 설정합니다.
model = Sequential()
model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1))
model.summary()

# 모델을 실행합니다.
model.compile(optimizer ='adam', loss = 'mean_squared_error')

# 100회 이상 결과가 향상되지 않으면 자동으로 중단되게끔 합니다.
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=100)

# 모델의 이름을 정합니다.
modelpath="/content/drive/MyDrive/kdt_2506/모두의딥러닝/data/miniproject2-3.keras"

# 최적화 모델을 업데이트하고 저장합니다.
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=0, save_best_only=True)

# 실행 관련 설정을 하는 부분입니다. 전체의 20%를 검증셋으로 설정합니다.
history = model.fit(X_train, y_train, validation_split=0.25, epochs=2000, batch_size=32, callbacks=[early_stopping_callback, checkpointer])
```

```python
from sklearn.metrics import r2_score

# 학습 후 평가
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R² score:", r2)
```

```python
# 예측 값과 실제 값, 실행 번호가 들어갈 빈 리스트를 만듭니다.
real_prices =[]
pred_prices = []
X_num = []

# 25개의 샘플을 뽑아 실제 값, 예측 값을 출력해 봅니다.
n_iter = 0
Y_prediction = model.predict(X_test).flatten()
for i in range(25):
    real = y_test.iloc[i]*stds + means
    prediction = Y_prediction[i]*stds + means
    print("실제팔로워: {:.2f}, 예상팔로워: {:.2f}".format(real, prediction))
    real_prices.append(real)
    pred_prices.append(prediction)
    n_iter = n_iter + 1
    X_num.append(n_iter)
```
